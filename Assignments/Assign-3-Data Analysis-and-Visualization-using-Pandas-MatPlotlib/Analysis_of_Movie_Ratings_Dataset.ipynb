```python
from datasets import load_dataset

dataset = load_dataset('ashraq/movielens_ratings')
df = dataset['train'].to_pandas()
```

# **Task: Top 10 most rated movies with a bar chart.**

```python
# Task: Most Rated Movies
most_rated_movies = df['movie_id'].value_counts().head(10)
most_rated_movies = most_rated_movies.reset_index()
most_rated_movies.columns = ['movie_id', 'rating_count']
```

```python
# Visualization: Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(most_rated_movies['movie_id'].astype(str), most_rated_movies['rating_count'], color='orange')
plt.title('Top 10 Most Rated Movies')
plt.xlabel('Movie ID')
plt.ylabel('Number of Ratings')
plt.grid(axis='y')
plt.show()
```

# **Task: Average Ratings per Movie**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Task: Average Ratings per Movie
average_ratings = df.groupby('movie_id')['rating'].mean()
```

```python
# Visualization: Histogram
plt.figure(figsize=(10, 6))
plt.hist(average_ratings, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Average Ratings per Movie')
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# **Task: Rating Range Visualization**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Assuming ratings is a NumPy array of ratings from the dataset
ratings = np.array([3.5, 4.0, 2.0, 5.0, 3.0, 4.5, 1.0, 2.5, 4.0, 5.0])

# Calculate min, max, and range
min_rating = np.min(ratings)
max_rating = np.max(ratings)
rating_range = max_rating - min_rating
```

```python
# Visualization: Line plot with markers for min and max ratings
plt.plot(ratings, color='blue', marker='o', label='Ratings')
plt.scatter([np.argmin(ratings)], [min_rating], color='red', label=f'Min: {min_rating}', zorder=5)
plt.scatter([np.argmax(ratings)], [max_rating], color='green', label=f'Max: {max_rating}', zorder=5)
plt.title(f'Ratings Range Visualization (Range: {rating_range})')
plt.xlabel('Index')
plt.ylabel('Rating')
plt.legend()
plt.grid(True)
plt.show()
```
