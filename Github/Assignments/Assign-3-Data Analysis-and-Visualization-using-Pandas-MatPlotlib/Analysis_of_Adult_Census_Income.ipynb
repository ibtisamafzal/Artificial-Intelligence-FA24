```python
from datasets import load_dataset

dataset = load_dataset('scikit-learn/adult-census-income')
df = dataset['train'].to_pandas()
```

# **Task: Analyze the distribution of income levels (<=50K vs >50K).**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Income Distribution Analysis
income_counts = df['income'].value_counts()
```

```python
# Visualization: Bar Plot
plt.figure(figsize=(8, 5))
income_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.legend(['<=50K'])
plt.title('Income Distribution')
plt.xlabel('Income Level')
plt.ylabel('Count')
plt.show()
```

# **Task: Age vs. Income Level**

```python
# Average Age by Income Level Analysis - Bar Plot
average_age = df.groupby('income')['age'].mean()

plt.figure(figsize=(9, 5))
average_age.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Average Age by Income Level')
plt.xlabel('Income Level')
plt.ylabel('Average Age')
plt.xticks(rotation=0)
plt.show()
```

# **Task: Education Level Impact**

```python
# Average Income by Education Level Analysis
education_income = df.groupby('education')['income'].value_counts(normalize=True).unstack()

# Visualization: Grouped Bar Chart
education_income.plot(kind='bar', figsize=(10, 5), color=['lightgreen', 'orange'])
plt.title('Average Income by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.legend(title='Income Level')
plt.show()
```
