```python
import numpy as np
```

**2D Array = Lowest and Highest Temperature of the day**

```python
temperatures = np.array([
    [27.5, 45.5],
    [36.5, 42.6],
    [33.7, 50.1],
    [30.5, 38.2]
    ])
print(temperatures)
```

```python
import matplotlib.pyplot as plt
```

```python
average_temp = np.mean(temperatures[:,0])
average_temp_2 = np.mean(temperatures[:,1])
```

```python
print(average_temp)
print(average_temp_2)
```

```python
overall_average_temp= (average_temp + average_temp_2)/2
```

```python
print (overall_average_temp)
```

```python
plt.figure(figsize=(10,5))
plt.plot(temperatures[:,0], marker="o", linestyle="--", color="blue",label="Low Temperature")
plt.plot(temperatures[:,1], marker="o", linestyle="--", color="orange",label="High Temperature")
plt.plot(overall_average_temp, linestyle="--", color="green",label="Average Temperature")
plt.title("Temperature Report of the week")
plt.ylabel("Temperature")
plt.xlabel("Days")
plt.legend()
plt.grid()
plt.show()
```
