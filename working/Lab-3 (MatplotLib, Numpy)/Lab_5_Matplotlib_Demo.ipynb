```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[0,5,10,15,20,25]

b=[1,2,3,4,5,6]
a=[1.1,2.1,3.1,4.1,5.1,6.1]

plt.plot(x,y, color="black", marker="o", label="x vs y")
plt.plot(a,b, color="black", marker="o", label="a vs b")
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line Plot")
plt.legend()
plt.fill_between(x,y,color="pink")
plt.fill_betweenx(a,b,color="blue")
plt.show()
```
